```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import nltk
```

```python
path = '/content/drive/MyDrive/Trendyol Data/'
```

```python
df_reviews = pd.read_csv(path+'reviews.csv')
```

```python
df_reviews['comment'] = df_reviews['comment'].apply(lambda x: re.sub('[,\.!?:()"]', '', x))
df_reviews['comment'] = df_reviews['comment'].apply(lambda x: x.lower())
df_reviews['comment'] = df_reviews['comment'].apply(lambda x: x.strip())
```

```python
stopwordlist = ['acaba', 'ama', 'aslında', 'az', 'bazı', 'belki', 'biri', 'birkaç', 'birşey', 'biz', 'bu', 'çok', 'çünkü', 'da', 'daha', 'de', 'defa', 'diye', 'eğer', 'en', 'gibi', 'hem', 'hep', 'hepsi', 'her', 'hiç', 'için', 'ile', 'ise', 'kez', 'ki', 'kim', 'mı', 'mu', 'mü', 'nasıl', 'ne', 'neden', 'nerde', 'nerede', 'nereye', 'niçin', 'niye', 'o', 'sanki', 'şey', 'siz', 'şu', 'tüm', 've', 'veya', 'ya', 'yani']
```

```python
def token(values):
    words = nltk.tokenize.word_tokenize(values)
    filtered_words = [word for word in words if word not in stopwordlist]
    not_stopword_doc = " ".join(filtered_words)
    return not_stopword_doc
```

```python
df_reviews['comment'] = df_reviews['comment'].apply(lambda x: token(x))
```

```python
df_rev_grp_sup.head()
```

```python
df_rev_grp_sup_dev = df_reviews.groupby('supplier_id')['rate','rate_bin_123','if_contains_iade'].std()
```

```python
df_rev_grp_sup_dev.rename(columns={'rate':'rate_dev','rate_bin_123':'rate_bin_123_dev','if_contains_iade':'if_contains_iade_dev'},inplace=True)
```

```python
df_mean_stdev_suppliers = df_rev_grp_sup.merge(df_rev_grp_sup_dev,how='inner',on='supplier_id')
```

```python
df_mean_stdev_suppliers.head()
```

```python
df_rev_grp_sup_prod.head(50)
```

```python
product_content_outlier_list = []
```

```python
for i in df_rev_grp_sup_prod.index.unique(level=0):
  supp_mean = df_mean_stdev_suppliers.loc[i]['rate']
  supp_stdev = df_mean_stdev_suppliers.loc[i]['rate_dev']
  threshold = supp_mean-supp_stdev
  for j in df_rev_grp_sup_prod.loc[i].index:
    if df_rev_grp_sup_prod.loc[i,j]['rate'] <threshold:
      product_content_outlier_list.append(j)
```
